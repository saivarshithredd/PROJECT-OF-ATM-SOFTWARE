CHECKING BALANCE IN ACCOUNT 

```python
print('PLEASE INSERT YOUR CARD')
pin=int(input('PLEASE ENTER YOUR PIN'))
balance=50000
if pin==1234:
    print("""
    1==BALANCE ENQUIRY
    2==WITHDRAW CASH
    3==FAST CASH
    4==EXIT""")
    option=int(input('PLEASE ENTER YOUR CHOICE'))
    if option==1:
        print('YOUR CURRENT BALANCE IS =',balance)
        print('PLEASE TAKE YOUR CARD BACK')
    if option==2:
        print('PLEASE SELECT ACCOUNT TYPE'
         """
         1==CURRENT
         2==SAVINGS""")
        account=int(input())
        if account==1 or 2:
            withdrawamount=int(input('PLEASE ENTER THE AMOUNT YOU WANT TO WITHDRAW'))
        else:
            print('PLEASE SELECT ONLY 1 OR 2')
        if withdrawamount<=50000:
            balance=balance-withdrawamount
            print(withdrawamount,'IS DEBITED FROM YOUR ACCOUNT')
            print('YOUR UPDATED CURRENT BALANCE IS =',balance)
            print('PLEASE COLLECT YOUR CASH')
            print('PLEASE TAKE YOUR CARD BACK')
        else:
            print("""INSUFFICIENT BALANCE
                              PLEASE ENTER THE AMOUNT WHICH IS AVAILABLE IN YOUR ACCOUNT SO YOU CAN WITHDRAW""")
    if option==3:
        print("""
    1==1000
    2==2000
    3==5000
    4==10000""")
        value=int(input('PLEASE ENTER YOUR CHOICE'))
        if value==1:
            balance=balance-1000
            print('1000 IS DEBITED FROM YOUR ACCOUNT')
            print('YOUR UPDATED BALANCE IS =',balance)
            print('PLEASE COLLECT YOUR CASH')
            print('PLEASE TAKE YOUR CARD BACK')
        if value==2:
            balance=balance-2000
            print('2000 IS DEBITED FROM YOUR ACCOUNT')
            print('YOUR UPDATED BALANCE IS =',balance)
            print('PLEASE COLLECT YOUR CASH')
            print('PLEASE TAKE YOUR CARD BACK')
        if value==3:
            balance=balance-5000
            print('5000 IS DEBITED FROM YOUR ACCOUNT')
            print('YOUR UPDATED BALANCE IS =',balance)
            print('PLEASE COLLECT YOUR CASH')
            print('PLEASE TAKE YOUR CARD BACK')
        if value==4:
            balance=balance-10000
            print('10000 IS DEBITED FROM YOUR ACCOUNT')
            print('YOUR UPDATED BALANCE IS =',balance)
            print('PLEASE COLLECT YOUR CASH')
            print('PLEASE TAKE YOUR CARD BACK')
    if option==4:
            print('EXITED')
            print('PLEASE TAKE YOUR CARD BACK')
else:
    print('WRONG PIN ENTERED RE-INSERT THE CARD AND TRY AGAIN ')

    
    

    

    
    
    



                             
    
    
```
